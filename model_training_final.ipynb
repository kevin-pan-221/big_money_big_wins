```python
#Data Processing
import pandas as pd
import numpy as np

#Model Training
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor #Random Forest 
from sklearn.neural_network import MLPRegressor #Neural Network
from sklearn.metrics import mean_squared_error, r2_score

#Plots
import matplotlib.pyplot as plt
```

```python
def add_std_deviation_column(df, column_name, new_column_name="Std_Devs_From_Mean"):
    mean_value = df[column_name].mean()
    std_dev = df[column_name].std()

    df[new_column_name] = (df[column_name] - mean_value) / std_dev
```

```python
#Reading in Data

# Serie A
serie_a_2020_2021 = pd.read_excel('Soccer Data/Serie A (done)/Serie A Stats - 2020-2023.xlsx', sheet_name='2020-2021')
add_std_deviation_column(serie_a_2020_2021, "Average Annual Payroll (Millions)")

serie_a_2021_2022 = pd.read_excel('Soccer Data/Serie A (done)/Serie A Stats - 2020-2023.xlsx', sheet_name='2021-2022')
add_std_deviation_column(serie_a_2021_2022, "Average Annual Payroll (Millions)")

serie_a_2022_2023 = pd.read_excel('Soccer Data/Serie A (done)/Serie A Stats - 2020-2023.xlsx', sheet_name='2022-2023')
add_std_deviation_column(serie_a_2022_2023, "Average Annual Payroll (Millions)")

combined_serie_a_2020_2023 = pd.concat([serie_a_2020_2021, serie_a_2021_2022, serie_a_2022_2023], ignore_index=True)

# Saudi Arabia
saudi_arabia_2023_2024 = pd.read_excel('Soccer Data/Saudi Arabia (done)/Saudi Arabia Stats 2023-2024.xlsx', sheet_name='2023-2024')
add_std_deviation_column(saudi_arabia_2023_2024, "Average Annual Payroll (Millions)")

# Premier League
prem_2020_2021 = pd.read_excel('Soccer Data/Premier League (done)/Premier League Stats - 2020-2023.xlsx', sheet_name='2020 - 2021')
add_std_deviation_column(prem_2020_2021, "Average Annual Payroll (Millions)")

prem_2021_2022 = pd.read_excel('Soccer Data/Premier League (done)/Premier League Stats - 2020-2023.xlsx', sheet_name='2021 - 2022')
add_std_deviation_column(prem_2021_2022, "Average Annual Payroll (Millions)")

prem_2022_2023 = pd.read_excel('Soccer Data/Premier League (done)/Premier League Stats - 2020-2023.xlsx', sheet_name='2022-2023')
add_std_deviation_column(prem_2022_2023, "Average Annual Payroll (Millions)")

combined_prem_2020_2023 = pd.concat([prem_2020_2021, prem_2021_2022, prem_2022_2023], ignore_index=True)

# MLS
mls_2020 = pd.read_excel('Soccer Data/MLS (done)/MLS Stats - 2020 - 2024 (USD)_.xlsx', sheet_name='2020')
add_std_deviation_column(mls_2020, "Average Annual Payroll (Millions)")

mls_2021 = pd.read_excel('Soccer Data/MLS (done)/MLS Stats - 2020 - 2024 (USD)_.xlsx', sheet_name='2021')
add_std_deviation_column(mls_2021, "Average Annual Payroll (Millions)")

mls_2022 = pd.read_excel('Soccer Data/MLS (done)/MLS Stats - 2020 - 2024 (USD)_.xlsx', sheet_name='2022')
add_std_deviation_column(mls_2022, "Average Annual Payroll (Millions)")

mls_2023 = pd.read_excel('Soccer Data/MLS (done)/MLS Stats - 2020 - 2024 (USD)_.xlsx', sheet_name='2023')
add_std_deviation_column(mls_2023, "Average Annual Payroll (Millions)")

mls_2024 = pd.read_excel('Soccer Data/MLS (done)/MLS Stats - 2020 - 2024 (USD)_.xlsx', sheet_name='2024')
add_std_deviation_column(mls_2024, "Average Annual Payroll (Millions)")

combined_mls_2020_2024 = pd.concat([mls_2020, mls_2021, mls_2022, mls_2023, mls_2024], ignore_index=True)

# Ligue 1
ligue_2020_2021 = pd.read_excel('Soccer Data/Ligue 1 (done)/Ligue 1 Stats - 2020-2023.xlsx', sheet_name='2020-2021')
add_std_deviation_column(ligue_2020_2021, "Average Annual Payroll (Millions)")

ligue_2021_2022 = pd.read_excel('Soccer Data/Ligue 1 (done)/Ligue 1 Stats - 2020-2023.xlsx', sheet_name='2021-2022')
add_std_deviation_column(ligue_2021_2022, "Average Annual Payroll (Millions)")

ligue_2022_2023 = pd.read_excel('Soccer Data/Ligue 1 (done)/Ligue 1 Stats - 2020-2023.xlsx', sheet_name='2022-2023')
add_std_deviation_column(ligue_2022_2023, "Average Annual Payroll (Millions)")

combined_ligue_2020_2023 = pd.concat([ligue_2020_2021, ligue_2021_2022, ligue_2022_2023], ignore_index=True)

# La Liga
liga_2020_2021 = pd.read_excel('Soccer Data/La Liga (done)/La Liga Stats 2020 - 2023.xlsx', sheet_name='2020-2021')
add_std_deviation_column(liga_2020_2021, "Average Annual Payroll (Millions)")

liga_2021_2022 = pd.read_excel('Soccer Data/La Liga (done)/La Liga Stats 2020 - 2023.xlsx', sheet_name='2021-2022')
add_std_deviation_column(liga_2021_2022, "Average Annual Payroll (Millions)")

liga_2022_2023 = pd.read_excel('Soccer Data/La Liga (done)/La Liga Stats 2020 - 2023.xlsx', sheet_name='2022-2023')
add_std_deviation_column(liga_2022_2023, "Average Annual Payroll (Millions)")

combined_liga_2020_2023 = pd.concat([liga_2020_2021, liga_2021_2022, liga_2022_2023], ignore_index=True)

# Bundesliga
bundes_2020_2021 = pd.read_excel('Soccer Data/Bundesliga (done)/Bundesliga Stats 2020 -2023.xlsx', sheet_name='2020 - 2021')
add_std_deviation_column(bundes_2020_2021, "Average Annual Payroll (Millions)")

bundes_2021_2022 = pd.read_excel('Soccer Data/Bundesliga (done)/Bundesliga Stats 2020 -2023.xlsx', sheet_name='2021 - 2022')
add_std_deviation_column(bundes_2021_2022, "Average Annual Payroll (Millions)")

bundes_2022_2023 = pd.read_excel('Soccer Data/Bundesliga (done)/Bundesliga Stats 2020 -2023.xlsx', sheet_name='2022 - 2023')
add_std_deviation_column(bundes_2022_2023, "Average Annual Payroll (Millions)")

combined_bundes_2020_2023 = pd.concat([bundes_2020_2021, bundes_2021_2022, bundes_2022_2023], ignore_index=True)

# Brazilian Serie A
brazil_2021 = pd.read_excel('Soccer Data/Brazilian Serie A (done)/Brazilian Serie A Stats - 2021 - 2023.xlsx', sheet_name='2021')
add_std_deviation_column(brazil_2021, "Total Annual Payroll (Millions)")

brazil_2022 = pd.read_excel('Soccer Data/Brazilian Serie A (done)/Brazilian Serie A Stats - 2021 - 2023.xlsx', sheet_name='2022')
add_std_deviation_column(brazil_2022, "Total Annual Payroll (Millions)")

brazil_2023 = pd.read_excel('Soccer Data/Brazilian Serie A (done)/Brazilian Serie A Stats - 2021 - 2023.xlsx', sheet_name='2023')
add_std_deviation_column(brazil_2023, "Total Annual Payroll (Millions)")

combined_brazil_2021_2023 = pd.concat([brazil_2021, brazil_2022, brazil_2023], ignore_index=True)

# Chinese Super League
csl_2019 = pd.read_excel('Chinese Super League Stats - 2019.xlsx')
add_std_deviation_column(csl_2019, "Average Annual Payroll (Millions)")
```

```python
# Function to add win_pct column
def add_win_pct_column(df):
    df["win_pct"] = (
        pd.to_numeric(df["W"], errors='coerce') 
        + 0.5 * pd.to_numeric(df["D"], errors='coerce')
    ) / pd.to_numeric(df["MP"], errors='coerce')
    return df

# Serie A
serie_a_2020_2021 = add_win_pct_column(serie_a_2020_2021)
serie_a_2021_2022 = add_win_pct_column(serie_a_2021_2022)
serie_a_2022_2023 = add_win_pct_column(serie_a_2022_2023)
combined_serie_a_2020_2023 = add_win_pct_column(combined_serie_a_2020_2023)

# Saudi Arabia
saudi_arabia_2023_2024 = add_win_pct_column(saudi_arabia_2023_2024)

# Premier League
prem_2020_2021 = add_win_pct_column(prem_2020_2021)
prem_2021_2022 = add_win_pct_column(prem_2021_2022)
prem_2022_2023 = add_win_pct_column(prem_2022_2023)
combined_prem_2020_2023 = add_win_pct_column(combined_prem_2020_2023)

# MLS
mls_2020 = add_win_pct_column(mls_2020)
mls_2021 = add_win_pct_column(mls_2021)
mls_2022 = add_win_pct_column(mls_2022)
mls_2023 = add_win_pct_column(mls_2023)
mls_2024 = add_win_pct_column(mls_2024)
combined_mls_2020_2024 = add_win_pct_column(combined_mls_2020_2024)

# Ligue 1
ligue_2020_2021 = add_win_pct_column(ligue_2020_2021)
ligue_2021_2022 = add_win_pct_column(ligue_2021_2022)
ligue_2022_2023 = add_win_pct_column(ligue_2022_2023)
combined_ligue_2020_2023 = add_win_pct_column(combined_ligue_2020_2023)

# La Liga
liga_2020_2021 = add_win_pct_column(liga_2020_2021)
liga_2021_2022 = add_win_pct_column(liga_2021_2022)
liga_2022_2023 = add_win_pct_column(liga_2022_2023)
combined_liga_2020_2023 = add_win_pct_column(combined_liga_2020_2023)

# Bundesliga
bundes_2020_2021 = add_win_pct_column(bundes_2020_2021)
bundes_2021_2022 = add_win_pct_column(bundes_2021_2022)
bundes_2022_2023 = add_win_pct_column(bundes_2022_2023)
combined_bundes_2020_2023 = add_win_pct_column(combined_bundes_2020_2023)

# Brazilian Serie A
brazil_2021 = add_win_pct_column(brazil_2021)
brazil_2022 = add_win_pct_column(brazil_2022)
brazil_2023 = add_win_pct_column(brazil_2023)
combined_brazil_2021_2023 = add_win_pct_column(combined_brazil_2021_2023)

# Chinese Super League
csl_2019 = add_win_pct_column(csl_2019)
```

```python
#Note: Chinese Super League & Brazilian Serie A do not have full data --> choose to leave out
#Premier league is left out for testing the modle
all_league_data = pd.concat(
    [combined_bundes_2020_2023,
     combined_liga_2020_2023,
     combined_ligue_2020_2023,
     combined_mls_2020_2024,
     combined_serie_a_2020_2023], ignore_index=True)
```

```python
# 48 features
# Reg_Rank - team rank in season
# Club - club name
# MP - matches played
# W - wins
# D - draws
# L - losses
# GF - goals scored
# GA - goals allowed
# GD - goal differential
# Pts - points
# Total Annual Payroll (Millions)
# Average Annual Payroll (Millions)
# Total Annual Payroll
# Average Annual Payroll
# Pl - Number of Players
# Age - Average Age weighted by Minutes Played
# Possession - Calculated as percentages of passes attempted
# MP.1 - Matches Played (same for all)
# Starts - Game starts (same for all)
# Min - Min played (same for all)
# 90s - Min played / 90 (same for all)
# Gls - Goals scored or allowed
# Assists - Assists
# G+A - Goals (scored or allowed) and Assists
# G-PK - Non-penalty Goals 
# PK - penalty kicks made
# PK - penalty kicks attempted
# CrdY - yellow cards
# CrdR - red cards
# xG - Expected goals
# npxG - non penalty expected goals
# xAG - expected assisted goals
# npxG + xAG - non penalty expected goals and expected assisted goals
# PrgC - progressive carries / dribbles
# PrgP - progressive passes
# Gls.1 - goals scored per 90 minutes
# Ast.1 - assists per 90 minutes
# G+A.1 - goals and assists per 90 minutes
# G-PK.1 - Non-penalty Goals per 90 mins
# G+A-PK - Non-penalty goals + assists per 90 minutes
# xG.1 - expected goals per 90 minutes
# xAG.1 - expected assisted goals per 90 minute
# xG+xAG - expected goals + assisted goals per 90 minute
# npxG.1 - non-penalty expected goals per 90 minutes
# npxG + xAG.1 - non-penalty expected goals + assisted goals per 90 minutes
# Std_Devs_From_Mean - Average Total Annual Payroll (Millions) Z score
# Playoff_Rank - only MLS
# win_pct - W + 0.5 D / MP
```

```python
#Drop Unnecessary Columns, or columns that are too descriptive in our opinion (i.e. pts)
cols_to_drop = ['Reg_Rank', 'Club', 'MP', 'W', 'D', 'L', 'GF', 'GA', 'GD', 'Pts', 'MP.1', 'Starts', 'Min', '90s']
all_league_data.drop(columns=cols_to_drop, inplace=True)
```

```python
all_league_data.shape
#34 features
```

```python
#Random Forest Model
# Step 1: Separate target and features
y = all_league_data['win_pct']  # The target variable
X = all_league_data.drop(columns=['win_pct'])  # Drop the target column

# Step 2: Identify numeric features and subset
numeric_cols = X.select_dtypes(include='number').columns
X_numeric = X[numeric_cols]

# Step 3: Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

# Step 4: Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Step 5: Train a Random Forest model on the training set
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Step 6: Evaluate on the test set
y_test_pred = rf_model.predict(X_test)
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)
print("Test MSE:", test_mse)
print("Test R²:", test_r2)
```

```python
#Feature Importances
importances = rf_model.feature_importances_
feature_importance_df = pd.DataFrame({'feature': numeric_cols, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

feature_importance_df['feature'] = feature_importance_df['feature'].replace({'Std_Devs_From_Mean': 'Standardized Average Annual Payroll (Millions)'})

plt.figure(figsize=(10,6))
plt.barh(feature_importance_df['feature'], feature_importance_df['importance'])
plt.gca().invert_yaxis()  # highest importance at top
plt.title('Feature Importances')
plt.xlabel('Importance Score')
plt.show()
```

```python
#Get the actual predictions & Residuals
y_pred = rf_model.predict(X_test)
residuals = y_test - y_pred
```

```python
#Predicted vs. Actual Wins
plt.figure(figsize=(8,6))
plt.scatter(y_test, y_pred, alpha=0.7)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.title('Random Forest Predicted vs. Actual Win %')
plt.xlabel('Actual Wins')
plt.ylabel('Predicted Wins')
plt.show()
```

```python
#Residuals vs. Actual
plt.figure(figsize=(8,6))
plt.scatter(y_test, residuals)
plt.hlines(y=0, xmin=min(y_test), xmax=max(y_test), color='red')
plt.title('Random Forest Residuals vs. Actual Values')
plt.xlabel('Actual Values')
plt.ylabel('Residual (Actual - Predicted)')
plt.show()
```

```python
#Neural Network
# Step 5: Train a Neural Network
#Originally had 2 layers with 64 nodes each - performed horribly
nn_model = MLPRegressor(hidden_laye r_sizes=(256, 128, 64, 32, 16),
                        activation='relu',
                        solver='adam',
                        random_state=42,
                        max_iter=1000)
nn_model.fit(X_train, y_train)

# Plot the training loss curve
plt.figure(figsize=(8,6))
plt.plot(nn_model.loss_curve_)
plt.title("Neural Network Training Loss Curve")
plt.xlabel("Iterations")
plt.ylabel("Loss")
plt.grid(True)
plt.show()

# Step 6: Evaluate on the test set
y_test_pred_nn = nn_model.predict(X_test)
test_mse_nn = mean_squared_error(y_test, y_test_pred_nn)
test_r2_nn = r2_score(y_test, y_test_pred_nn)

#print("Neural Network Test MSE:", test_mse_nn)
#print("Neural Network Test R²:", test_r2_nn)
```
