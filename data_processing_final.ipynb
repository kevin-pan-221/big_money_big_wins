```python
def add_std_deviation_column(df, column_name, new_column_name="Std_Devs_From_Mean"):
    mean_value = df[column_name].mean()
    std_dev = df[column_name].std()

    df[new_column_name] = (df[column_name] - mean_value) / std_dev
```

```python
def calculate_regression(df, column_x, column_y):
    # Drop rows with NaN values in the selected columns
    clean_data = df[[column_x, column_y]].dropna()
    
    x = clean_data[column_x]
    y = clean_data[column_y]
    
    # Calculate the Pearson correlation coefficient and R^2
    r, _ = pearsonr(x, y)
    r_squared = r ** 2
    
    return {"r": r, "R^2": r_squared}
```

```python
def plot_regression_results(results, title):
    seasons = list(results.keys())
    r_values = [results[season]['r'] for season in seasons]
    r_squared_values = [results[season]['R^2'] for season in seasons]
    
    # Plot r values
    plt.figure(figsize=(10, 6))
    plt.plot(seasons, r_values, marker='o', linestyle='-', color='dodgerblue', label='r (Correlation Coefficient)')
    plt.plot(seasons, r_squared_values, marker='o', linestyle='-', color='darkorange', label='R² (Coefficient of Determination)')

    plt.ylim(-1, 1)
    
    plt.title(title, fontsize=16, fontweight='bold', color='darkslategray')
    plt.xlabel("Seasons", fontsize=14, fontweight='bold')
    plt.ylabel("Value", fontsize=14, fontweight='bold')
    plt.legend(fontsize=12)
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

def calculate_and_plot_regression(df, column_x, column_y, title=None, x_label=None, y_label=None, invert_y=False):
    # Clean data
    clean_data = df[[column_x, column_y]].dropna()
    x = clean_data[column_x]
    y = clean_data[column_y]
    
    # Calculate Pearson correlation and R^2
    r, _ = pearsonr(x, y)
    if invert_y:
        r = -1 * r
    r_squared = r ** 2
    
    # Calculate regression line
    slope, intercept = np.polyfit(x, y, 1)
    regression_line = slope * x + intercept

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, color='dodgerblue', alpha=0.8, edgecolor='k', s=80)
    
    # Add regression line and legend with r and R^2
    plt.plot(x, regression_line, color='darkblue', linewidth=2,
             label=f"Regression Line\n(y = {slope:.2f}x + {intercept:.2f})\n"
                   f"r = {r:.2f}, R² = {r_squared:.2f}")

    # Add x and y axis labels
    plt.xlabel(x_label if x_label else column_x, fontsize=14, fontweight='bold')
    plt.ylabel(y_label if y_label else column_y, fontsize=14, fontweight='bold')
    
    # Add custom or default title
    graph_title = title if title else f"Regression Analysis: {column_x} vs {column_y}"
    plt.title(graph_title, fontsize=16, fontweight='bold', color='darkslategray')
    
    # Add legend and grid
    plt.legend(fontsize=12)
    plt.grid(color='gray', linestyle='--', linewidth=0.5, alpha=0.7)
    plt.tick_params(axis='both', which='major', labelsize=12)

    # Invert y-axis if requested
    if invert_y:
        plt.gca().invert_yaxis()

    plt.tight_layout()
    plt.show()
    
    #return {"r": r, "R^2": r_squared}
```

```python


# Calculate correlation between Wage and Performance
def calculate_correlation(wage_df, perf_df):
    """
    Calculate the correlation (r) and R^2 between spending (wage) and performance.
    
    Parameters:
        wage_df (pd.DataFrame): DataFrame containing wage data (spending).
        perf_df (pd.DataFrame): DataFrame containing performance data.
        
    Returns:
        pd.DataFrame: Correlation results (r and R^2) for each league.
    """
    leagues = wage_df.columns[0:]  # Exclude 'Year'
    results = []

    for league in leagues:
        wage = wage_df[league]
        perf = perf_df[league]
        
        # Drop NaN values to align data
        wage_clean, perf_clean = wage.dropna(), perf.dropna()
        
        # Calculate correlation and R^2
        r, _ = pearsonr(wage_clean, perf_clean)
        r = -1 * r
        r_squared = r ** 2
        
        # Store results
        results.append({"League": league, "r": r, "R^2": r_squared})
    
    # Convert results to a DataFrame
    results_df = pd.DataFrame(results)
    return results_df

# Calculate correlation
correlation_results = calculate_correlation(CL_wage, CL_perf)

# Display results
print("Correlation between Wage and Performance:")
print(correlation_results)
```

```python
def plot_correlation_bar_chart(correlation_df, title="Correlation Between Wage and Performance"):
    # Set up plot
    leagues = correlation_df['League']
    r_values = correlation_df['r']
    r_squared_values = correlation_df['R^2']
    
    x = range(len(leagues))  # X-axis positions

    plt.figure(figsize=(10, 6))
    
    # Plot bars for r and R²
    plt.bar(x, r_values, width=0.4, label="r (Correlation Coefficient)", color='dodgerblue')
    plt.bar([i + 0.4 for i in x], r_squared_values, width=0.4, label="R² (Coefficient of Determination)", color='orange')
    
    # Add labels and title
    plt.xticks([i + 0.2 for i in x], leagues, rotation=45, ha="right")
    plt.ylabel("Values", fontsize=12, fontweight='bold')
    plt.title(title, fontsize=14, fontweight='bold', color='darkslategray')
    
    # Add legend and grid
    plt.legend(fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.ylim(-1, 1)
    plt.tight_layout()
    plt.show()

# Example correlation results DataFrame (your data)
correlation_results = pd.DataFrame({
    "League": ["Premier League", "Bundesliga", "La Liga", "Serie A", "Ligue 1"],
    "r": [0.476312, 0.256098, -0.720533, 0.141919, 0.130827],
    "R^2": [0.226873, 0.066048, 0.519167, 0.020141, 0.017116]
})

# Call the function to plot
plot_correlation_bar_chart(correlation_results, "Big 5 Leagues Correlation Between Wage and Performance in Champions League (2016-2023)")
```
