```python
# Create a unified 'Club' column by taking the first non-null value between 'Club' and 'Team'
all_league_data['GF'] = all_league_data['GF'].combine_first(all_league_data['Gls'])

# Now that we have a unified column, we can drop the 'Team' column
all_league_data.drop(columns=['Gls'], inplace=True)
all_league_data.drop(columns=['Playoff_Rank'], inplace=True)
all_league_data.drop(columns=['Pts'], inplace=True)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Assume all_league_data is your DataFrame

# Step 1: Separate target and features
y = all_league_data['W']  # The target variable is the number of wins
X = all_league_data.drop(columns=['W'])  # Drop the target from the features

# (Optional) If you have categorical variables, you may need to encode them:
# For example, if 'Club' is categorical:
# X = pd.get_dummies(X, columns=['Club'], drop_first=True)

# Step 2: Identify numeric features and subset
numeric_cols = X.select_dtypes(include='number').columns
X_numeric = X[numeric_cols]

# Step 3: Standard Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_numeric)

# Step 4: Create train-validation-test splits
# First split off the test set
X_train_val, X_test, y_train_val, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)

# Then split the train_val into train and validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val, test_size=0.25, random_state=42
)
# Now:
# - Training set: 60% of data
# - Validation set: 20% of data
# - Test set: 20% of data

# Step 5: Train a Random Forest model on the training set
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Step 6: Evaluate on the validation set (for hyperparameter tuning)
y_val_pred = model.predict(X_val)
val_mse = mean_squared_error(y_val, y_val_pred)
val_r2 = r2_score(y_val, y_val_pred)
print("Validation MSE:", val_mse)
print("Validation R²:", val_r2)

# (Optional) If needed, perform hyperparameter tuning here
# e.g., try different n_estimators, max_depth, etc., and re-check on X_val, y_val

# After finalizing the model/hyperparameters,
# retrain the model on the combined (train+val) set for the final evaluation:
model_final = RandomForestRegressor(n_estimators=100, random_state=42)
model_final.fit(X_train_val, y_train_val)

# Evaluate on the test set
y_test_pred = model_final.predict(X_test)
test_mse = mean_squared_error(y_test, y_test_pred)
test_r2 = r2_score(y_test, y_test_pred)
print("Test MSE:", test_mse)
print("Test R²:", test_r2)
```
